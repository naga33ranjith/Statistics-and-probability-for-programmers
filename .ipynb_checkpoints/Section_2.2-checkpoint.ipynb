```python
import numpy as np
import pandas as pd
from pandas import DataFrame,Series
import statistics 
```

```python
#2.15w
NBA=pd.read_excel('NBACost2013.xlsx')
```

```python
NBA.sort_values(by='Cost($)',inplace=True)
```

```python
def f(data):
    return DataFrame({"Relative Frequency": (data['Cost($)'])/(NBA['Cost($)'].sum()),
                      "Percentage Distribution":(data['Cost($)'])/(NBA['Cost($)'].sum())*100
                     })
Distribution=f(NBA)
```

```python
pd.merge(NBA,Distribution,left_index=True,right_index=True)
```

```python
#2.16
utility=pd.read_excel("utility.xlsx")
utility.shape
```

```python
utility.info()
```

```python
def f(x):
    
    return DataFrame({"Relative Frequency":(x['Utility Charge'])/(x['Utility Charge'].sum()),
                      "Percentage Distribution":(x['Utility Charge'])/(x['Utility Charge'].sum())*100,
                      'Cumulative Distribution':((x['Utility Charge'])/(x['Utility Charge'].sum())*100).cumsum()
                     })
dist=f(utility)
```

```python
pd.merge(utility,dist,left_index=True,right_index=True)
    
    
```
